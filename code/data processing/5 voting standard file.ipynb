```python
def hex_to_binary(hex_data):
    # Convert hexadecimal to 8-bit binary
    return bin(int(hex_data, 16))[2:].zfill(8)

def binary_to_hex(binary_data):
    # Convert binary back to hexadecimal
    return hex(int(binary_data, 2))[2:].upper().zfill(2)

def vote_on_bits(*binary_data):
    # Vote on each bit
    length = len(binary_data[0])
    voted_result = ''

    for i in range(length):
        # Count the number of 1s in each bit position
        ones_count = sum(b[i] == '1' for b in binary_data)

        # If the number of 1s is greater than or equal to 3, the result is 1; otherwise, it's 0
        voted_result += '1' if ones_count >= 3 else '0'

    return voted_result

# Read files and perform conversion
with open("21.txt", 'r') as f1, open("22.txt", 'r') as f2, open("23.txt", 'r') as f3, open("24.txt", 'r') as f4, open("25.txt", 'r') as f5:
    hex_data1 = f1.read().strip().split()
    hex_data2 = f2.read().strip().split()
    hex_data3 = f3.read().strip().split()
    hex_data4 = f4.read().strip().split()
    hex_data5 = f5.read().strip().split()

# Convert each hexadecimal number to binary
binary_data1 = [hex_to_binary(hex_num) for hex_num in hex_data1]
binary_data2 = [hex_to_binary(hex_num) for hex_num in hex_data2]
binary_data3 = [hex_to_binary(hex_num) for hex_num in hex_data3]
binary_data4 = [hex_to_binary(hex_num) for hex_num in hex_data4]
binary_data5 = [hex_to_binary(hex_num) for hex_num in hex_data5]

# Voting result in hexadecimal
voted_results = []
for i in range(len(binary_data1)):
    voted_result = vote_on_bits(binary_data1[i], binary_data2[i], binary_data3[i], binary_data4[i], binary_data5[i])
    voted_results.append(binary_to_hex(voted_result))

# Save the voted hexadecimal numbers to a file
with open("biaozhun5t5.txt", 'w') as f:
    f.write(' '.join(voted_results))

# Read the standard document and the documents to compare
with open("biaozhun5t5.txt", 'r') as f_standard:
    standard_hex = f_standard.read().strip().split()

# Read the standard document and the documents to compare
with open("21.txt", 'r') as f_standard1:
    hex_data1 = f_standard1.read().strip().split()

# Read the standard document and the documents to compare
with open("22.txt", 'r') as f_standard2:
    hex_data2 = f_standard2.read().strip().split()

# Read the standard document and the documents to compare
with open("23.txt", 'r') as f_standard3:
    hex_data3 = f_standard3.read().strip().split()

# Read the standard document and the documents to compare
with open("24.txt", 'r') as f_standard4:
    hex_data4 = f_standard4.read().strip().split()

# Read the standard document and the documents to compare
with open("25.txt", 'r') as f_standard5:
    hex_data5 = f_standard5.read().strip().split()

# Convert hexadecimal to binary
standard_binary = ''.join([hex_to_binary(hex_num) for hex_num in standard_hex])
comparison_binary1 = ''.join([hex_to_binary(hex_num) for hex_num in hex_data1])
comparison_binary2 = ''.join([hex_to_binary(hex_num) for hex_num in hex_data2])
comparison_binary3 = ''.join([hex_to_binary(hex_num) for hex_num in hex_data3])
comparison_binary4 = ''.join([hex_to_binary(hex_num) for hex_num in hex_data4])
comparison_binary5 = ''.join([hex_to_binary(hex_num) for hex_num in hex_data5])

def calculate_reliability(standard_binary, comparison_binaries):
    # Ensure there are five comparison_binary arrays in comparison_binaries
    if len(comparison_binaries) != 5:
        raise ValueError("Five comparison_binary arrays are required")

    # Calculate reliability
    total_bits = len(standard_binary)
    same_bits = 0
    for i in range(total_bits):
        if all(standard_binary[i] == comparison_binary[i] for comparison_binary in comparison_binaries):
            same_bits += 1
    return same_bits / total_bits

def calculate_reliability(standard_binary, comparison_binary):
    # Calculate reliability
    total_bits = len(standard_binary)
    same_bits = sum(s == c for s, c in zip(standard_binary, comparison_binary))
    return same_bits / total_bits

reliability1 = calculate_reliability(standard_binary, comparison_binary1)
reliability2 = calculate_reliability(standard_binary, comparison_binary2)
reliability3 = calculate_reliability(standard_binary, comparison_binary3)
reliability4 = calculate_reliability(standard_binary, comparison_binary4)
reliability5 = calculate_reliability(standard_binary, comparison_binary5)

reliability = (reliability1 + reliability2 + reliability3 + reliability4 + reliability5) / 5
```
